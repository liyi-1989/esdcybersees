# Step 1. Load the GCM data

```r
# ipython nbconvert script.ipynb --to slides --post serve
library(MASS)
library(ncdf4)
get_gcm=function(mylon,mylat){
  cat("In position: ",mylon,mylat,"\n")
  files=list.files("./gcm")
  pr=NULL
  for(i in files){
    cat("Read in file ",i,"\n")
    ncin = nc_open(paste0("./gcm/",i))
    lon = ncvar_get(ncin,"lon")
    lat = ncvar_get(ncin,"lat")
    Pr=ncvar_get(ncin,"pr")
    pr=c(pr,Pr[((mylon-0.1)<lon)&(lon<(mylon+0.1)),((mylat-0.1)<lat)&(lat<(mylat+0.1)),])
  }
  prm=matrix(pr,365,length(pr)/365)
  return(prm)
}

X=get_gcm(-116.25+180,43) # kg/m^2/s = mm/hr *3600
colnames(X)=1975:2005; rownames(X)=1:365
X=X[,5:29] * 3600 * 24 # mm # 365 days * 25 years
MON=1:365; names(MON)=c(rep("JAN",31),rep("FEB",28),rep("MAR",31),rep("APR",30),rep("MAY",31),rep("JUN",30),rep("JUL",31),rep("AUG",31),rep("SEP",30),rep("OCT",31),rep("NOV",30),rep("DEC",31))
X1=X[MON[names(MON)=="JAN"],] # 31 days * 25 years
X1m=apply(X1,2,sum) # 1 month * 25 years 
```

```r
# 25 years: 1979 - 2003
dim(X) # GCM pr 25 years(365 whole year)
dim(X1) # GCM pr 25 years(31 Jan)
X1m # GCM pr 25 years(Jan Accumulation) in mm
```

```r
plot(ecdf(X1m))
```

# Step 2. Load the Observation data

```r
Y=read.csv("obs_avg.csv",header = F)
colnames(Y)=1979:2003; rownames(Y)=1:365
Y1=Y[MON[names(MON)=="JAN"],] # 31 days * 25 years
Y1m=apply(Y1,2,sum) # 1 month * 25 years 
Y1m
```

# Step 3: Empirical CDF used for Quantile Mapping

```r
plot(sort(X1m),rank(sort(X1m))/length(X1m),type="o",xlim=c(0,max(Y1m)),xlab="Preicipitation",ylab="ECDF",col="blue")
lines(sort(Y1m),rank(sort(Y1m))/length(Y1m),type="o",col="red")
legend("top",c("GCM","Obs"),lty=1,pch=1,col=c("blue","red"))
```

# Step 4: Quantile Mapping

```r
Q1=quantile(Y1m,ecdf(X1m)(X1m))
plot(1979:2003,X1m,type="o",ylim=c(0,max(Y1m)),col="blue",xlab="Years",ylab="Precipitation(mm)",main="Bias Correction")
lines(1979:2003,Y1m,type="o",col="red")
lines(1979:2003,Q1,type="o",col="green")
legend("top",c("GCM","Obs","Quantile Mapping"),lty=1,pch=1,col=c("blue","red","green"))
```

# Step 5: Copula Modification

Essentially: 

- Given Data
    - $X$: GCM
    - $Y$: Obs
- Task
    - provide GCM predict $x_t$
    - what is $y_t$, the correction?

$$y_t=argmax_{Obs}f(Obs|GCM)=argmax_{y}f(y|x_t)$$

So with copula $f(x,y)=c(u,v)f(x)f(y)$, $U=F_{GCM}(X)$, $V=F_{Obs}(Y)$, we have

$$y_t=argmax_{y}f(y|x_t)=argmax_{y}\frac{f(x_t,y)}{f(x_t)}=argmax_{y}c(u_t,v)f(y)$$

```r
kde1d_point=function(x0,x_given){
  density(x_given, from=x0, to=x0, n=1)$y
}

kde2d_point=function(x0,y0,x_given,y_given){
  kde2d(x_given,y_given,lims=c(x0,x0,y0,y0),n=1)$z
}

L=length(X1m)
plot(rank(X1m)/L,rank(Y1m)/L,xlab="GCM",ylab="Obs",main="25 years Precipitation Data")
```

```r
f=kde2d(rank(X1m)/L,rank(Y1m)/L)
contour(f,main="Joint Copula Density")
```

```r
plot(density(Y1m),main="Obs Marginal Density")
```

```r
# Function to find the max (y) of the conditional density based on the data
max_cond=function(x_fix,x_given,y_given){
  ygrid=seq(from=min(y_given),to=max(y_given),length=100)
  yt=ygrid
  u=rank(x_given)/25
  v=rank(y_given)/25
  for(i in 1:length(ygrid)){
    yt[i]=kde2d_point(rank(c(x_fix,x_given))[1]/26,rank(c(ygrid[i],y_given))[1]/26,u,v)*kde1d_point(x_fix,x_given)
  }
  ygrid[which.max(yt)] #ygrid[yt==max(yt)] #plot(ygrid,yt,type="o")
}

C1=X1m
for(i in 1:length(X1m)){
  C1[i]=max_cond(X1m[i],X1m,Y1m)
}
```

# Step 6: Compare the Bias Correction Results

```r
plot(1979:2003,X1m,type="o",ylim=c(0,max(Y1m)),col="blue",xlab="Years",ylab="Precipitation(mm)",main="Bias Correction")
lines(1979:2003,Y1m,type="o",col="red")
lines(1979:2003,Q1,type="o",col="green")
lines(1979:2003,C1,type="o",col="black")
legend("top",c("GCM","Obs","Quantile Mapping","Nonparametric Copula"),lty=1,pch=1,col=c("blue","red","green","black"))
```

```r
sqrt(mean((Y1m-Q1)^2)) # RMSE for Quantile Mapping
sqrt(mean((Y1m-C1)^2)) # RMSE for Copula
```

# Repeat the same thing in July

```r
X7=X[MON[names(MON)=="JUL"],] # 31 days * 25 years
X7m=apply(X7,2,sum) # 1 month * 25 years 
Y7=Y[MON[names(MON)=="JUL"],] # 31 days * 25 years
Y7m=apply(Y7,2,sum) # 1 month * 25 years 
Q7=quantile(Y7m,ecdf(X7m)(X7m))
C7=X7m
for(i in 1:length(X7m)){
  C7[i]=max_cond(X7m[i],X7m,Y7m)
}
```

```r
plot(1979:2003,X7m,type="o",ylim=c(0,max(Y7m)),col="blue",xlab="Years",ylab="Precipitation(mm)",main="Bias Correction")
lines(1979:2003,Y7m,type="o",col="red")
lines(1979:2003,Q7,type="o",col="green")
lines(1979:2003,C7,type="o",col="black")
legend("top",c("GCM","Obs","Quantile Mapping","Nonparametric Copula"),lty=1,pch=1,col=c("blue","red","green","black"))

sqrt(mean((Y7m-Q7)^2)) # RMSE for Quantile Mapping
sqrt(mean((Y7m-C7)^2)) # RMSE for Copula
```
